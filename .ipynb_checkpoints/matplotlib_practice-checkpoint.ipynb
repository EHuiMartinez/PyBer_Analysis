```python
%matplotlib inline
```

```python
#Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot [matlab/pyplot method]
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot with ax.plt() [object-oriented method]
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot with ax.plt() [2nd method]

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fares($)")

#Set the y limit between 0 and 45.
plt.ylim(0, 45)

#Create a title.
plt.title("PyBer Fare by Month")

#Add the legend.
plt.legend()

plt.show()
```

```python
#Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fares($)")

#Set the y limit between 0 and 45.
plt.ylim(0, 45)
#plt.xlim(0, 5)

#Create a title.
plt.title("PyBer Fare by Month")

#Add a grid.
plt.grid()

#Add the legend.
plt.legend()

#plt.figure(figsize=(20, 10))
#plt.subplots(figsize=(20,10))

plt.show()
```

```python
#Skill Drill (object-oriented method)

#Change the line color to green.
#Change the marker to a diamond.
#Change the line width to 2 points.

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')

#Add a legend for the city of Boston.
ax.legend()

#Add a title and axes labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

#Add a y-axis limit.
ax.set_ylim(0, 45)

#Add grid lines.
ax.grid()


plt.show()
```

```python
#Create the plot (bar charts)
plt.bar(x_axis, y_axis, color="green", label='Boston')

#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title.
plt.title("Pyber Fare by Month")

#Add the legend.
plt.legend()

plt.show()
```

```python
#Create the plot (horizontal bar chart) "gca= get current axes"
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()

plt.show()
```

```python
#Skill Drill 

#Create the plot (horizontal bar chart) "gca= get current axes"
#Change the bar colors to magenta.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')

#Add a legend for the city of Boston.
plt.legend()

#Add a title and axes labels.
plt.title("Pyber Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")

#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()

plt.show()
```

```python
#Create scatter plot ("o" changes plot line to scatter plot)
plt.plot(x_axis, y_axis, 'o')

plt.show()
```

```python
#Create scatter plot
plt.scatter(x_axis, y_axis)

plt.show()

```

```python
#Skill Drill (scatter plot)

#Change the color of the markers to red.
plt.scatter(y_axis, x_axis, color="red", label='Chicago')

#Add a legend for the city of Chicago.
plt.legend()

#Add a title and axes labels.
plt.title("Pyber Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")

#Switch the axis so that the Fare($) data is on the x-axis.

#Add a limit for the x-axis data.
plt.xlim(0, 50)

#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()

plt.show()
```

```python
#Create bubble chart (size added (3rd value) to scatter)

#increase marker size
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)

plt.show()
```

```python
#Create bubble chart (size added (3rd value) to scatter) - refactored

#increase marker size (refactored, using Python's 'list comprehension')
#new_list = [expression for item in list if conditional]
    
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])

plt.show()
```

```python
#Create bubble chart (size added (3rd value) to scatter) [obj-oriented]
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)

plt.show()
```

```python
#Skill Drill: #Create bubble chart (size added (3rd value) to scatter) [obj-oriented]

#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Make the color 20% transparent.
#Add a black edge color to the circles.
#Make the linewidth of the circles 2 points.


fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s= [i*5 for i in y_axis], color="skyblue", alpha=0.8, edgecolor="black", label='Boston')

#Add a legend for the city of Boston.
ax.legend()

#Add a title
ax.set_title("PyBer Fare by Month")

#Switch the axis so that the Fare($) data is on the x-axis.
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')

#Add a limit for the x-axis data.
ax.set_xlim(0, 50)

#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()

plt.show()
```

```python
#Create pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#Create explode values in pie chart for emphasis
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors, 
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

```python
#Create pie charts (obj-oriented)
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
#Skill drill: create pie charts (obj-oriented)

#Add a percentage to one decimal place to each wedge of the pie.
#Increase the figure size to 8x8.
#Explode the two highest percentage months.
#Add a shadow.
#Add a start angle so that January is at the top.
#Reverse the order so that the month order is in a clockwise direction.
#Add new colors of your own choosing or use the colors from the previous pie chart.


#Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#Create explode values in pie chart for emphasis
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, 
        explode=explode_values, 
        colors=colors, 
        labels=x_axis,
        shadow=True,
        autopct='%.1f%%',
        startangle=90,
        counterclock=False)

plt.show()

```
